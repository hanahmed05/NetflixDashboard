```python
import pandas as pd
import numpy as np

# Visualization
import plotly.express as px
import plotly.graph_objects as go

# Display options
pd.set_option("display.max_columns", None)

# Load data
df = pd.read_csv("netflix_titles.csv")
```

```python
df = df.copy()

# Remove duplicate rows
df = df.drop_duplicates()

# Convert dates
# Remove leading/trailing whitespace first
df["date_added"] = df["date_added"].astype(str).str.strip()

# Convert to datetime
df["date_added"] = pd.to_datetime(
    df["date_added"],
    errors="coerce"
)
# Fill missing text values
text_columns = [
    "director",
    "cast",
    "country",
    "rating"
]

for col in text_columns:
    df[col] = df[col].fillna("Unknown")

# Drop rows missing genres
df = df.dropna(subset=["listed_in"])

# Create year/month columns
df["year_added"] = df["date_added"].dt.year
df["month_added"] = df["date_added"].dt.month_name()

df.head()
```

```python
# Extract duration number (e.g., 90 from "90 min" or 2 from "2 Seasons")
df["duration_number"] = (
    df["duration"]
    .str.extract(r"(\d+)")
    .astype(float)
)

df["duration_number"] = (
    df["duration"]
    .str.extract(r"(\d+)")
    .astype(float)
    .astype("Int64")  
)

# Simplify genres (keep only the first listed genre)
df["main_genre"] = (
    df["listed_in"]
    .str.split(",")
    .str[0]
    .str.strip()
)

df.head()
```
